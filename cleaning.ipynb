```python
import numpy as np
import pandas as pd
```

```python
flight_df = pd.read_feather('data/flights')
flight_test_df = pd.read_feather('data/flight_test')
passenger_df = pd.read_feather('data/passenger')
fuel_consumption_df = pd.read_feather('data/fuel_consumption')
```

```python
col_to_drop_flight = ['dep_time','wheels_off','wheels_on','arr_time','cancelled','cancellation_code','diverted',
                    'dup','actual_elapsed_time','air_time','carrier_delay','weather_delay','nas_delay','security_delay',
                    'late_aircraft_delay','first_dep_time','total_add_gtime','longest_add_gtime','no_name']
v1_flight_df = flight_df.drop(col_to_drop_flight,axis=1)
#column full of ones, not useful
v1_flight_df = v1_flight_df.drop('flights',axis=1) 
v1_flight_df.head()
```

```python
col_to_drop_flight_test = ['mkt_carrier','branded_code_share','op_carrier_fl_num','origin','dest','dup','flights']
v1_flight_test_df = flight_test_df.drop(col_to_drop_flight_test,axis=1)
v1_flight_test_df.head()
```

```python
cols_to_drop = ['dest_country','carrier_group','carrier_group_new','origin','origin_country','dest','dest_country','dest_city_market_id',
                'data_source','distance_group','aircraft_config','aircraft_group','origin_city_market_id','ramp_to_ramp','carrier']
v1_passenger_df = passenger_df.drop(cols_to_drop,axis=1)
v1_passenger_df.head()
```

```python
fuel_drop_col = ['unique_carrier','sdomt_gallons','satl_gallons','spac_gallons','slat_gallons','sint_gallons',
                'ts_gallons','tdomt_gallons','tint_gallons','sdomt_cost','satl_cost','spac_cost','slat_cost',
                'sint_cost','ts_cost','tdomt_cost','tint_cost']
v1_fuel_consumption_df = fuel_consumption_df.drop(fuel_drop_col,axis=1)
v1_fuel_consumption_df.dropna(inplace=True)
```

```python
v1_fuel_consumption_df.reset_index(inplace=True)
v1_fuel_consumption_df = v1_fuel_consumption_df.drop('index',axis=1)
```

```python
v1_flight_df.to_feather('data/v1_clean_flight')
v1_flight_test_df.to_feather('data/v1_clean_flight_test')
v1_fuel_consumption_df.to_feather('data/v1_clean_fuel_consumption')
v1_passenger_df.to_feather('data/v1_clean_passenger')
```
